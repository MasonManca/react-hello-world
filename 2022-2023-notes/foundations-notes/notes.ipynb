##### Ancient Egypt:

- Geography
    * Nile is the backbone of civilization ( Gods of fertility )
        - Floods, silt, irrigation; great agriculture
        - Transportation of goods and ideas
    * Desert allowed them to be isolated

- Ancient Egyption Timeline:
    * Based on unity or disunity
- Predynastic Period - summerians starting civilization
    * Rock drawings and identifiable egyption culture
    * The dead faces west since the east is for life
- Protodynastic - Emergence of tombs, monumental brick
- Early Dynastic - Unified lower and upper Egypt
    * First stable monarchy
    * God level status of pharoahs 
- Old Kingdom - Pyramids are being built
    * Trade in Nubia and Palestine 
    - First Intermediate Period
        * Rise of nobles and theives
- Middle Kingdom Egypt - rise in geographic domination
    * Development of Mythology
    * External invasion 
    - Second Intermediate Period 
        * Ideas of a standing army
- New Kingdom - warrior pharoah
    * Ramsay II, King Tut
    * Job is to conquer and protect
    - Third intermediate period
        * Collapse of bronze age civilizations
        * Chaos everywhere
- Late Period - more invasions 
    * Persions, Asyrians
- Greco Roman Period (332 BCE) - Alexander the Great conquers Egypt
    * General Tolmy is left in charge of Egypt
    * Cleopatra is the last Tolmy
